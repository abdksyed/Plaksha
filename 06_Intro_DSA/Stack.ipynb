```python
# Implementing Stacks using Arrays 

class ArrayStack:
    
    # Constructor for ArrayStack class
    def __init__(self):
        # your code here
        self.initialSize = 5
        self.currentPosition = -1
        self.stack = [None] * self.initialSize
        

    # Push: Adds a new element at the back of the list
    def push(self, element):
        # your code here
        self.currentPosition += 1
        if (self.currentPosition >= self.initialSize):
            self.initialSize += self.initialSize
            self.newStack = [None] * self.initialSize
            for elementIndex in range(self.currentPosition):
                self.newStack[elementIndex] = self.stack[elementIndex]
            self.stack = self.newStack
        self.stack[self.currentPosition] =  element


    # Pop: Deletes the element at the last and returns the value of it
    def pop(self):
        # your code here
        if self.currentPosition < 0:
            return -1
        if self.stack[self.currentPosition] != None:
            deletedElement = self.stack[self.currentPosition]
            self.stack[self.currentPosition] = None
            self.currentPosition -= 1
            return deletedElement
        return -1


    # Returns the size of the stack
    def size(self):
        # your code here
        return self.currentPosition + 1


    # Return the element at the top of the stack without removing it
    def top(self):
        if self.stack[self.currentPosition] != None:
            return self.stack[self.currentPosition]
        return -1


    # Returns true is stack is empty, False if not
    def isEmpty(self):
        # your code here
        if self.currentPosition >= 0:
            return False
        return True


    def printIsEmpty(self):
        print("\nStack is Empty\n") if self.isEmpty() else print("\nStack is not Empty\n")


    def printStack(self):
        print(self.stack)

stk = ArrayStack()

stk.size()

stk.push(2)

stk.push(3)

stk.push(4)

stk.push(5)

stk.push (6)

stk.printStack()

stk.push(7)

stk.push(8)

stk.top()

stk.size()

stk.isEmpty()

stk.printStack()

stk.pop()

stk.printStack()

stk.printIsEmpty()
```

```python
# Implementing Stacks Linked Lists

# Node class for the individual nodes
class Node:

    # constructor for Node class
    def __init__(self, data):
        # your code here
        self.data = data
        self.next = None

# Manager class to link the nodes and manage the overall list
class LinkedListStack:

    # constructor for LinkedListStack class
    def __init__(self):
        # your code here
        self.stackSize = 0

    # Push: Adds a new element at the back of the list
    def push(self, data):
        # your code here
        if (self.stackSize == 0):
            self.headNode = Node(data)
        else:
            newNode = Node(data)
            newNode.next = self.headNode
            self.headNode = newNode
        
        self.stackSize += 1

    # Pop: Deletes the element at the last and returns the value of it
    def pop(self):
        # your code here
        if(self.headNode != None and self.stackSize > 0):
            currentNode = self.headNode
            self.headNode = self.headNode.next
            headData = currentNode.data
            self.stackSize -= 1
            del currentNode
            return headData
        return None


    # Returns the size of the stack
    def size(self):
        # your code here
        return self.stackSize


    # Return the element at the top of the stack without removing it
    def top(self): 
        # your code here
        return self.headNode.data


    # Return true is stack is empty, False if not
    def isEmpty(self):
        # your code here
        if self.stackSize == 0:
            return True
        return False


    def printIsEmpty(self):
        print("\nStack is Empty\n") if self.isEmpty() else print("\nStack is not Empty\n")

    # Reverses the stack
    def reverseList(self):
        # your code here
        if(self.stackSize > 1):
            previousNode = self.headNode
            currentNode = previousNode.next
            previousNode.next = None
            nextNode = currentNode.next
            while(currentNode != None):
                nextNode = currentNode.next
                currentNode.next = previousNode
                previousNode = currentNode
                currentNode = nextNode
            self.headNode = previousNode

    # Print the linked list in reverse direction
    def printReverseLinkedList(self, currentNode):
        if(currentNode != None):
            self.printReverseLinkedList(currentNode.next)
            print(currentNode.data)

    # Print data stored in the stack
    def printStack(self):
        # your code here
        currentNode = self.headNode
        self.printReverseLinkedList(currentNode)

stk = LinkedListStack()

stk.push(2)

stk.push(3)

stk.push(4)

stk.push(5)

stk.printStack()
# print(stk.pop())
# print(stk.pop())
# print(stk.pop())
# print(stk.pop())

# stk.printIsEmpty()

# stk.push(1)

# stk.printStack()

stk.reverseList()

stk.printStack()
```

```python
import random as random
from time import perf_counter

# the program to be timed here
def calculateTime(stackObject, iterations):
    start = perf_counter()
    timeList  = []
    for i in range(iterations):
        number = random.randint(1,100)
        if (type(stackObject) == deque):
            stackObject.append(number)
        else:
            stackObject.push(number)
        if (i//100 == 0):
            end = perf_counter()
            timeList.append(end-start)
    return timeList

iterations = 10_00_000
arrayStack = ArrayStack()
arrayTimeList = calculateTime(arrayStack, iterations)
doublingArrayStack = DoublingArrayStack()
doublingArrayStackTimeList = calculateTime(doublingArrayStack, iterations)
linkedListStack = LinkedListStack()
linkedListTimeList = calculateTime(linkedListStack, iterations) 
internalStack = deque()
internalStackTimeList = calculateTime(internalStack, iterations)

plt.plot(arrayTimeList)
plt.plot(doublingArrayStackTimeList)
plt.plot(linkedListTimeList)
plt.plot(internalStackTimeList)
plt.legend(["Constant Increase", "Doubling Size", "Linked List", "Internal Stack"])
plt.title("Comparison of Different Stack Implementation")
```
